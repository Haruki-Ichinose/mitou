```python
import pandas as pd

path = "/workspace/data/raw/StatsAllGroup.csv"
df = pd.read_csv(path, encoding="cp932")

df["date_"] = pd.to_datetime(df["date_"]).dt.normalize()

daily = (
    df.groupby(["athlete_id", "date_"])
      .agg({
          "athlete_name": "first",
          "total_distance": "sum",
          "total_player_load": "sum",
          "mean_heart_rate": "mean",
          "max_heart_rate": "max",
          "min_heart_rate": "min",
          "max_vel": "max",
      })
      .reset_index()
)

print(daily.shape)
print(daily.head())
```
